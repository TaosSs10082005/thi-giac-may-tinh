```python
import cv2
import numpy as np
import matplotlib.pyplot as Hung
```

```python
img = cv2.imread('image.jpg')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

stop_data = cv2.CascadeClassifier('stop_data.xml')

found = stop_data.detectMultiScale(img_gray, minSize=(20, 20))

for (x, y, w, h) in found:
    cv2.rectangle(img_rgb, (x, y), (x+w, y+h), (0,255,0), 3)

Hung.imshow(img_rgb)
Hung.title("Stop Detection")
Hung.axis('off')
Hung.show()
```

```python
img2 = cv2.imread('image2.jpg')
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
rgb2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')

faces = face_cascade.detectMultiScale(gray2, 1.3, 5)

for (x, y, w, h) in faces:
    cv2.rectangle(rgb2, (x,y), (x+w,y+h), (255,0,0), 3)

Hung.imshow(rgb2)
Hung.title("Face Detection")
Hung.axis('off')
Hung.show()
```

```python
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')

img3 = cv2.imread('image2.jpg')
gray3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
rgb3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)

eyes = eye_cascade.detectMultiScale(gray3, 1.3, 5)

for (x, y, w, h) in eyes:
    cv2.rectangle(rgb3, (x,y), (x+w,y+h), (0,255,0), 2)

Hung.imshow(rgb3)
Hung.title("Eye Detection")
Hung.axis('off')
Hung.show()
```

```python
smile_cascade = cv2.CascadeClassifier('haarcascade_smile.xml')

img4 = cv2.imread('image2.jpg')
gray4 = cv2.cvtColor(img4, cv2.COLOR_BGR2GRAY)
rgb4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)

smiles = smile_cascade.detectMultiScale(gray4, 1.8, 20)

for (x, y, w, h) in smiles:
    cv2.rectangle(rgb4, (x,y), (x+w,y+h), (255,255,0), 2)

Hung.imshow(rgb4)
Hung.title("Smile Detection")
Hung.axis('off')
Hung.show()
```
